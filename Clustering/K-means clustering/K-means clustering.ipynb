## Libraries

```python
#Necessary libraries
## Basic lbraries
import pandas as pd
import numpy as np

## Dataset
from sklearn.datasets import make_blobs

## Visualization and model evaluation
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline

## Model
from sklearn.cluster import KMeans
```

## Data

```python
#Create artificial data set
raw_data = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8)
```

### Data visualization

```python
#Visualize the data

plt.scatter(raw_data[0][:,0], raw_data[0][:,1])

plt.scatter(raw_data[0][:,0], raw_data[0][:,1], c=raw_data[1])
```

## Model deployment

```python
#Build and train the model



model = KMeans(n_clusters=4)

model.fit(raw_data[0])

#See the predictions

model.labels_

model.cluster_centers_
```

## Model evaluation

```python
#PLot the predictions against the original data set

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))

ax1.set_title('Our Model')

ax1.scatter(raw_data[0][:,0], raw_data[0][:,1],c=model.labels_)

ax2.set_title('Original Data')

ax2.scatter(raw_data[0][:,0], raw_data[0][:,1],c=raw_data[1])
```

NB: Since K-means is a cluster algorithm, it has no numerical or percent calculation of the accuracy of the model. K-means is used to find clusters in data. So how can we know that this result is the best? We can try to use a couple clustering algorithm at once and compare all of the end result.
